```python
df = pd.read_csv("clt_cleaned.csv")
clt_cleaned = df

dataframe1 = clt_cleaned.sort_values('Monthl_ Income', ascending=False)
dataframe2 = clt_cleaned.sort_values('Monthly_Home_Cost', ascending=False)

fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.bar(dataframe1.index, dataframe1['Monthl_ Income'])
ax1.set_xticklabels(dataframe1.index, rotation=60, horizontalalignment='right', fontsize='12')

ax1.set_title('Monthly Income', fontsize=22)
ax1.set_ylabel('Monthly Income')
ax1.set_xlabel('County')

ax2.bar(dataframe2.index, dataframe2['Monthly_Home_Cost'])
ax2.set_xticklabels(dataframe2.index, rotation=60, horizontalalignment='right', fontsize='12')

ax2.set_title('Monthly Home Cost', fontsize=22)
ax2.set_ylabel('Monthly Home Cost')
ax2.set_xlabel('County')

plt.style.use("fivethirtyeight")

plt.show()
```
