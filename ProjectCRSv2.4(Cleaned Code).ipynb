```python
# Dependencies
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
%matplotlib inline
```

```python
# File to Load
month_home_cost = "Average Median Monthly Housing Cost Cleaned Up test.csv"
month_caro_income = "Median NC and SC Monthly Income Cleaned Up.csv"
prop_value = "Median Property Value by County Cleaned Up.csv" 
sale_price = "Median Sales Price CSV Cleaned up AVG Per Year.csv" 
sale_inv = "Sales Inventory Cleaned Up AVG per Year.csv" 
year_income = "US County Median Yearly Income Cleaned Up.csv"
county_pop = "US County Population Cleaned Up.csv" 
```

```python
# Read  and store into Pandas data frame
month_home_cost_data = pd.read_csv(month_home_cost)
month_caro_income_data = pd.read_csv(month_caro_income)
prop_value_data = pd.read_csv(prop_value, encoding = "ISO-8859-1")
sale_price_data = pd.read_csv(sale_price)
sale_inv_data = pd.read_csv(sale_inv)
year_income_data = pd.read_csv(year_income, encoding = "ISO-8859-1")
county_pop_data = pd.read_csv(county_pop, encoding = "ISO-8859-1")
```

```python
# Collecting a list of all columns within the DataFrame
county_pop_data.columns
```

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame
trim_month_home_cost1 = month_home_cost_data[['County','2013', '2014', '2015', '2016', '2017', '2018']]
trim_month_caro_inc1 = month_caro_income_data[['County','2013', '2014', '2015', '2016', '2017', '2018']]
trim_prop_value1 = prop_value_data[['County','2013', '2014', '2015', '2016', '2017', '2018']]
trim_sale_price1 = sale_price_data[['County','2013', '2014', '2015', '2016', '2017', '2018']]
trim_sale_inv1 = sale_inv_data[['County','2013', '2014', '2015', '2016', '2017', '2018']]
trim_year_inc1 = year_income_data[['County','2013', '2014', '2015', '2016', '2017', '2018']]
trim_county_pop1 = county_pop_data[['County','2013', '2014', '2015', '2016', '2017', '2018']]
```

```python
trim_county_pop1
```

```python
# Move Year to Column and Set Values to what data is, this will cut down on # of Columns
trim_month_home_cost=trim_month_home_cost1.melt(id_vars=["County"], 
              var_name="Year", value_name="Mo. Home Cost")
trim_month_caro_inc=trim_month_caro_inc1.melt(id_vars=["County"], 
              var_name="Year", value_name="Mo. Caro Inc")
trim_prop_value=trim_prop_value1.melt(id_vars=["County"], 
              var_name="Year", value_name="Avg Prop Value")
trim_sale_price=trim_sale_price1.melt(id_vars=["County"], 
              var_name="Year", value_name="Avg Sale Price")
trim_sale_inv=trim_sale_inv1.melt(id_vars=["County"], 
              var_name="Year", value_name="Homes For Sale")
trim_year_inc=trim_year_inc1.melt(id_vars=["County"], 
              var_name="Year", value_name="Avg Yearly Inc")
trim_county_pop=trim_county_pop1.melt(id_vars=["County"], 
              var_name="Year", value_name="Population")
```

```python
cols = list(trim_sale_price.columns)
print(cols)
```

```python
unfilter_data = trim_month_home_cost.append([trim_month_caro_inc, trim_prop_value, trim_sale_price, trim_sale_inv, trim_year_inc, trim_county_pop])
unfilter_data
```

```python
#Trim Counties
clt_area= unfilter_data.loc[(unfilter_data["County"] == "York County") | 
	(unfilter_data["County"] == "Lancaster County") |
	(unfilter_data["County"] == "Union County") |
	(unfilter_data["County"] == "Mecklenburg County") |
	(unfilter_data["County"] == "Gaston County") |
	(unfilter_data["County"] == "Licoln County") |
	(unfilter_data["County"] == "Irdell County") |
	(unfilter_data["County"] == "Cabarrus County"), :]
clt_area.set_index(['County', 'Year'])
```

```python
clt_area.columns
```

```python
dataframe1 = clt_area.sort_values('Mo. Caro Inc', ascending=False)
dataframe2 = clt_area.sort_values('Mo. Home Cost', ascending=False)

fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.bar(dataframe1.index, dataframe1['Mo. Caro Inc'])
ax1.set_xticklabels(dataframe1.index, rotation=60, horizontalalignment='right', fontsize='12')

ax1.set_title('Monthly Income', fontsize=22)
ax1.set_ylabel('Monthly Income')
ax1.set_xlabel('County')

ax2.bar(dataframe2.index, dataframe2['Mo. Home Cost'])
ax2.set_xticklabels(dataframe2.index, rotation=60, horizontalalignment='right', fontsize='12')

ax2.set_title('Monthly Home Cost', fontsize=22)
ax2.set_ylabel('Monthly Home Cost')

plt.style.use("fivethirtyeight")

plt.show()
```
